The [`KFold`](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html) object in SciKit Learn tells the cross validation object (see below) how to split up the data:

### Coding this up


````python
# List to store R² scores for each fold
scores = []

# Initialize Linear Regression model
lr = LinearRegression()

# Initialize StandardScaler for feature standardization
s = StandardScaler()

# Perform K-Fold cross-validation
for train_index, test_index in kf.split(X):
    # Split data into training and testing sets for this fold
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y[train_index], y[test_index]
    
    # Standardize training features
    X_train_s = s.fit_transform(X_train)
    
    # Fit Linear Regression on scaled training data
    lr.fit(X_train_s, y_train)
    
    # Standardize test features using the same scaler
    X_test_s = s.transform(X_test)
    
    # Make predictions on the test set
    y_pred = lr.predict(X_test_s)

    # Compute R² score for this fold
    score = r2_score(y_test.values, y_pred)
    
    # Append the score to the list
    scores.append(score)
````

### `Pipeline` and `cross_val_predict`


`Pipeline` lets you chain together multiple operators on your data that both have a `fit` method.


````python
# StandardScaler will normalize features 
# so that each has mean = 0 and variance = 1.
s = StandardScaler()

# LinearRegression is the base regression model from sklearn
# that minimizes the residual sum of squares.
lr = LinearRegression()
````

### Combine multiple processing steps into a `Pipeline`

A pipeline contains a series of steps, where a step is ("name of step", actual_model). The "name of step" string is only used to help you identify which step you are on, and to allow you to specify parameters at that step.  


````python
# Create a pipeline that first standardizes the data then fits a linear regression model
estimator = Pipeline([("scaler", s),
                      ("regression", lr)])
````

### `cross_val_predict`

[`cross_val_predict`](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_predict.html) is a function that does K-fold cross validation for us, appropriately fitting and transforming at every step of the way.


````python
kf
````

````python
# Get cross-validated predictions
predictions = cross_val_predict(estimator, X, y, cv=kf)
````

````python
# Compute the R² score between true values and cross-validated predictions
r2_score(y, predictions)
````

````python
# Mean R² score across the folds
np.mean(scores) # almost identical!
````

Note that `cross_val_predict` doesn't use the same model for all steps; the predictions for each row are made when that row is in the validation set. We really have the collected results of 3 (i.e. `kf.num_splits`) different models. 

When we are done, `estimator` is still not fitted. If we want to predict on _new_ data, we still have to train our `estimator`. 


## Hyperparameter tuning


### Definition

**Hyperparameter tuning** involves using cross validation (or train-test split) to determine which hyperparameters are most likely to generate a model that _generalizes_ well outside of your sample.

### Mechanics

We can generate an exponentially spaces range of values using the numpy [`geomspace`](https://docs.scipy.org/doc/numpy/reference/generated/numpy.geomspace.html#numpy.geomspace) function.

```python
np.geomspace(1, 1000, num=4)
```

produces:

```
array([    1.,    10.,   100.,  1000.])
```

Use this function to generate a list of length 10 called `alphas` for hyperparameter tuning:


````python
# Create 10 values spaced geometrically (logarithmically) 
# between 1e-9 (very small) and 1e0 (which is 1)
alphas = np.geomspace(1e-9, 1e0, num=10)
alphas
````

The code below tunes the `alpha` hyperparameter for Lasso regression.


````python
# Evaluate Lasso regression for each alpha using cross-validation
scores = []
coefs = []
for alpha in alphas:
    las = Lasso(alpha=alpha, max_iter=100000)
    
    estimator = Pipeline([
        ("scaler", s),
        ("lasso_regression", las)])

    predictions = cross_val_predict(estimator, X, y, cv = kf)
    
    score = r2_score(y, predictions)
    
    scores.append(score)
````

````python
# Pair each alpha value with its corresponding R² score
# alphas → list of tested regularization strengths
# scores → list of model performance (R²) at each alpha
# list(zip(...)) will create a list of tuples: (alpha, score)
list(zip(alphas,scores))
````

````python
# Fit a Lasso regression model with a very small alpha = 1e-6
# Since alpha is close to zero, the Lasso penalty is almost negligible
# → the model behaves very similarly to ordinary least squares (OLS)
# → coefficients will look close to the unregularized regression coefficients
Lasso(alpha=1e-6).fit(X, y).coef_
````

````python
# Fit a Lasso regression model with alpha = 1.0
# Alpha controls the strength of the regularization (higher alpha → more shrinkage)
# Some coefficients may be set exactly to zero (feature selection effect)
Lasso(alpha=1.0).fit(X, y).coef_
````

````python

plt.figure(figsize=(10,6))
plt.semilogx(alphas, scores, '-o')
plt.xlabel('$\\alpha$')
plt.ylabel('$R^2$');
````

### Exercise

Add `PolynomialFeatures` to this `Pipeline`, and re-run the cross validation with the `PolynomialFeatures` added.

**Hint #1:** pipelines process input from first to last. Think about the order that it would make sense to add Polynomial Features to the data in sequence and add them in the appropriate place in the pipeline.

**Hint #2:** you should see a significant increase in cross validation accuracy from doing this


````python
# Create polynomial features of degree 2
pf = PolynomialFeatures(degree=2)

# Store R² scores for each alpha
scores = []

# Define a range of alpha values (Lasso regularization strength) on a logarithmic scale
alphas = np.geomspace(0.001, 10, 5)

# Loop through each alpha value
for alpha in alphas:
    # Initialize Lasso regression with given alpha
    las = Lasso(alpha=alpha, max_iter=100000)
    
    # Build pipeline: polynomial features -> scaling -> Lasso regression
    estimator = Pipeline([
        ("make_higher_degree", pf),
        ("scaler", s),
        ("lasso_regression", las)
    ])
    
    # Perform cross-validated predictions
    predictions = cross_val_predict(estimator, X, y, cv=kf)
    
    # Compute R² score for current alpha
    score = r2_score(y, predictions)
    
    # Store the score
    scores.append(score)

# Display collected scores
scores
````

## Grid Search CV


To do cross-validation, we used two techniques:
- use `KFolds` and manually create a loop to do cross-validation
- use `cross_val_predict` and `score` to get a cross-valiated score in a couple of lines.

To do hyper-parameter tuning, we see a general pattern:
- use `cross_val_predict` and `score` in a manually written loop over hyperparemeters, then select the best one.

Perhaps not surprisingly, there is a function that does this for us -- `GridSearchCV`


````python
from sklearn.model_selection import GridSearchCV

# Same estimator as before
estimator = Pipeline([("polynomial_features", PolynomialFeatures()),
        ("scaler", StandardScaler()),
        ("ridge_regression", Ridge())])

params = {
    'polynomial_features__degree': [1, 2, 3],
    'ridge_regression__alpha': np.geomspace(4, 20, 20)
}

grid = GridSearchCV(estimator, params, cv=kf)
````

````python
# Fit the grid search
grid.fit(X, y)
````

````python
# Display the best cross-validated score and the best set of hyperparameters found by GridSearchCV
grid.best_score_, grid.best_params_
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import (StandardScaler, 
                                   PolynomialFeatures)
from scipy.stats.mstats import normaltest
from scipy.stats import boxcox
from scipy.special import inv_boxcox
# Load Boston Housing dataset
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_data = pd.read_csv(file_name)
# Define target (y) and features (X)
lr = LinearRegression()
y_col = "MEDV"
X = boston_data.drop(y_col, axis=1)
y = boston_data[y_col]
# Expand features with polynomial features (degree=2, no bias term)
pf = PolynomialFeatures(degree=2, include_bias=False)
X_pf = pf.fit_transform(X)
# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_pf, y, test_size=0.3, 
                                                    random_state=72018)
# Standardize features
s = StandardScaler()
X_train_s = s.fit_transform(X_train)
# Apply Box-Cox transformation to target (y_train)
bc_result = boxcox(y_train)
y_train_bc = bc_result[0]
lam = bc_result[1]
# Train Linear Regression on transformed y
lr.fit(X_train_s, y_train_bc)
# Transform test features and predict
X_test_s = s.transform(X_test)
y_pred_bc = lr.predict(X_test_s)
# Inverse transform predictions back to original scale
y_pred_tran = inv_boxcox(y_pred_bc, lam)
# Evaluate R² score (true values first, predictions second)
print(r2_score(y_pred_tran,y_test)) #RES 0.848052537981275

lr = LinearRegression()
lr.fit(X_train_s,y_train)
lr_pred = lr.predict(X_test_s)
r2_score(lr_pred,y_test) #RES 0.8667029116056716
````

## Summary

1. We can manually generate folds by using `KFolds`
2. We can get a score using `cross_val_predict(X, y, cv=KFoldObject_or_integer)`. 
   This will produce the out-of-bag prediction for each row.
3. When doing hyperparameter selection, we should be optimizing on out-of-bag scores. This means either using `cross_val_predict` in a loop, or ....
4. .... use `GridSearchCV`. GridSearchCV takes a model (or pipeline) and a dictionary of parameters to scan over. It finds the hyperparameter set that has the best out-of-sample score on all the parameters, and calls that it's "best estimator". It then retrains on all data with the "best" hyper-parameters.

### Extensions

Here are some additional items to keep in mind:
* There is a `RandomSearchCV` that tries random combination of model parameters. This can be helpful if you have a prohibitive number of combinations to test them all exhaustively.
* KFolds will randomly select rows to be in the training and test folds. There are other methods (such as `StratifiedKFolds` and `GroupKFold`, which are useful when you need more control over how the data is split (e.g. to prevent data leakage). You can create these specialized objects and pass them to the `cv` argument of `GridSearchCV`.
